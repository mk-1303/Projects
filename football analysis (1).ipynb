```python
Total_goals=df['Goals'].sum()
print(Total_goals)   #Total no of goals scored in 2020-2021 season
```

```python
Penalty_goals=df['Penalty_Goals'].sum()
print(Penalty_goals)   #Total no of penalty goals
```

```python
Penalty_Attempted=df['Penalty_Attempted'].sum()
print(Penalty_Attempted) #Total no of penalty attempted
```

```python
Non_penalty_goals=Total_goals-Penalty_goals
print(Non_penalty_goals)  #Total no of non penalty goals
```

```python
Missed_penalty=Penalty_Attempted-Penalty_goals
print(Missed_penalty)  #Total no of penalty missed
```
